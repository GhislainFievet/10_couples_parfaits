```python
import pandas as pd
import itertools
```

```python
femmes = ['anissa','carla','dita','eva','kellyn','lea_mary','maissane','melanie','orelie','safia']
hommes = ['adrien','alcaraz','allan','anthony','bryan','khephren','krys','marin','seby_daddy','vincent']
```

```python
def nombreDeMatchs(h_couples,h_couples_ceremonie):
    compteur = 0
    for femme_solution in h_couples:
        homme_solution = h_couples[femme_solution]
        homme_ceremonie = h_couples_ceremonie[femme_solution]
        if homme_solution == homme_ceremonie:
            compteur += 1
    return compteur
```

```python
def SolutionValide(h_couples):
    if h_couples['kellyn'] == 'alcaraz':
        return False
    if h_couples['maissane'] == 'krys':
        return False
    if h_couples['safia'] == 'vincent':
        return False
    if h_couples['eva'] == 'anthony':
        return False
    if h_couples['dita'] == 'khephren':
        return False
    if h_couples['dita'] == 'alcaraz':
        return False
    if h_couples['safia'] != 'krys':
        return False
    if h_couples['anissa'] != 'marin':
        return False
    if h_couples['maissane'] == 'vincent':
        return False
    if h_couples['dita'] == 'bryan':
        return False
    
    couples_ceremonie_1 = {
    'anissa':'alcaraz',
    'carla':'allan',
    'eva':'seby_daddy',
    'dita':'bryan',
    'kellyn':'adrien',
    'lea_mary':'anthony',
    'maissane':'marin',
    'melanie':'krys',
    'orelie':'vincent',
    'safia':'khephren'
    }
    
    if nombreDeMatchs(h_couples,couples_ceremonie_1) != 2:
        return False
        
    couples_ceremonie_2 = {
    'anissa':'alcaraz',
    'carla':'allan',
    'dita':'bryan',
    'eva':'anthony',
    'kellyn':'adrien',
    'lea_mary':'seby_daddy',
    'maissane':'vincent',
    'melanie':'marin',
    'orelie':'khephren',
    'safia':'krys'
    }
    
    if nombreDeMatchs(h_couples,couples_ceremonie_2) != 2:
        return False

    couples_ceremonie_3 = {
    'anissa':'krys',
    'carla':'marin',
    'dita':'anthony',
    'eva':'seby_daddy',
    'kellyn':'bryan',
    'lea_mary':'allan',
    'maissane':'vincent',
    'melanie':'adrien',
    'orelie':'alcaraz',
    'safia':'khephren'
    }
    
    if nombreDeMatchs(h_couples,couples_ceremonie_3) != 2:
        return False
    
    couples_ceremonie_4 = {
    'anissa':'krys',
    'carla':'anthony',
    'dita':'alcaraz',
    'eva':'khephren',
    'kellyn':'adrien',
    'lea_mary':'seby_daddy',
    'maissane':'bryan',
    'melanie':'marin',
    'orelie':'vincent',
    'safia':'allan'
    }
    
    if nombreDeMatchs(h_couples,couples_ceremonie_4) != 2:
        return False
    
    couples_ceremonie_5 = {
    'anissa':'marin',
    'carla':'khephren',
    'dita':'anthony',
    'eva':'allan',
    'kellyn':'bryan',
    'lea_mary':'seby_daddy',
    'maissane':'vincent',
    'melanie':'adrien',
    'orelie':'alcaraz',
    'safia':'krys'
    }
    
    if nombreDeMatchs(h_couples,couples_ceremonie_5) != 3:
        return False
    
    couples_ceremonie_6 = {
    'anissa':'marin',
    'carla':'bryan',
    'dita':'anthony',
    'eva':'alcaraz',
    'kellyn':'khephren',
    'lea_mary':'seby_daddy',
    'maissane':'allan',
    'melanie':'adrien',
    'orelie':'vincent',
    'safia':'krys'
    }
    
    if nombreDeMatchs(h_couples,couples_ceremonie_6) != 4:
        return False
    
    couples_ceremonie_7 = {
    'anissa':'marin',
    'carla':'allan',
    'dita':'anthony',
    'eva':'seby_daddy',
    'kellyn':'adrien',
    'lea_mary':'bryan',
    'maissane':'khephren',
    'melanie':'alcaraz',
    'orelie':'vincent',
    'safia':'krys'
    }
    
    if nombreDeMatchs(h_couples,couples_ceremonie_7) != 9:
        return False
    return True
    
```

```python
permutations_object = itertools.permutations(hommes)

permutations_liste = list(permutations_object)

solutions_valides = []

for solution in permutations_liste:
    h_couples = {}
    femme_index = 0
    for homme in solution:
        h_couples[femmes[femme_index]] = homme
        femme_index += 1
    if SolutionValide(h_couples):
        solutions_valides.append(h_couples.copy())
```
